# importaciones y peticiones a la pagina

```python
import re
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
url = 'https://biblioteca.japan-paw.net/0:/Anime/D/[Madness%20Subs]%20Date%20A%20Live%20S3%20[BD%201080p]/[Madness%20Subs]%20Date%20A%20Live%20S3%20[BD%201080p].html'
url2 = 'https://biblioteca.japan-paw.net/0:/Anime/G/[MKNF]%20Grisaia%20no%20Kajitsu%20[BD%201080p]/[MKNF]%20Grisaia%20no%20Kajitsu%20[BD%201080p].html'

cabeceras = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'
}

r = requests.get(url2, headers=cabeceras)
sopita = bs(r.text, 'html.parser')
```

# Imprimimos el contenido

```python
print(sopita.prettify())
```

# Asi obtenemos el Titulo

```python
titulo = sopita.find("title").text
print(titulo)

# titulo2 = sopita.select('title')
# print(titulo2)
```

# Asi obtenemos los nombres de los capitulos

```python
nomb_caps = sopita.select('a.enlaces p')
for i in nomb_caps:
  print(i.text)

# print('\n\n' + (40*'++') + '\n\n')

# nomb_caps2 = sopita.select('.enlaces p')
# for j in nomb_caps2:
#   print(j.text)
```

# Asi obtenemos los Enlaces

```python
lnks = sopita.select('a.enlaces')
actual_lnks = [links['href'] for links in lnks]

actual_lnks
```

# Limpiamos con REGEX

```python
links_sanitos = []

patron = re.compile(r'https?://biblioteca.+\.m[kp]{1,1}[v4]{1,1}')
for link in actual_lnks:
  link = re.findall(patron, link)
  links_sanitos.append(link)

for i in links_sanitos:
  print(i)
```

#Informacion Faltante

```python
ruta = "/content/drive/MyDrive/colab/csv's/eluniversal_2021_12_20_articulos.csv"
#3 paso, relleanr elementos faltantes


missing_title_mask = eluniversal['title'].isna()

missing_titles = (
    eluniversal[missing_title_mask]['url']
    .str.extract(r'(?P<missing_titles>[^/]+$)')
    .applymap( lambda title: title.split('-') )
    .applymap( lambda title_word_list: ' '.join(title_word_list) )
)

missing_titles
```

#Limpiando detallles adicionales

```python
#4. añadir uid a las filas usando hashes
import hashlib

uids = (
    eluniversal
    .apply( lambda row: hashlib.md5(bytes(row['url'].encode())) ,axis=1 )
    .apply( lambda hash_object: hash_object.hexdigest() )
)

eluniversal['uid'] = uids
eluniversal.set_index('uid', inplace=True)
eluniversal
```

```python
#5. Eliminar saltos de linea del body
stripped_body = (
    eluniversal
    .apply( lambda row: row['body'], axis=1 )
    .apply( lambda body: list(body) )
    .apply( lambda letters: list(map(lambda letter: letter.replace('\n',''), letters)) )
    .apply( lambda letters_list: ''.join(letters_list) )
)

stripped_body
```
